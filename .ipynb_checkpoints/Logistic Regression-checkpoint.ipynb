```python
import numpy as np 
import pandas as pd
from pandas import Series, DataFrame

import math

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

from sklearn.linear_model import LogisticRegression 
from sklearn.cross_validation import train_test_split

from sklearn import metrics

import statsmodels.api as sm
```

```python
def logistic(t):
    return 1.0 / (1 + math.exp((-1.0)*t))

t = np.linspace(-6, 6, 500)

y = np.array([logistic(ele) for ele in t])

plt.plot(t, y)
plt.title('Logistic Function')
```

```python
df = sm.datasets.fair.load_pandas().data
```

```python
df.head()
```

```python
def affair_check(x):
    if x != 0:
        return 1 
    else:
        return 0 
    
```

```python
df['Had_Affair'] = df['affairs'].apply(affair_check)
```

```python
df.head()
```

```python
df.groupby('Had_Affair').mean()
```

```python
df.head()
```

```python
sns.factorplot('age', data=df, hue='Had_Affair', kind='count', palette='coolwarm')
```

```python
sns.factorplot('yrs_married', data=df, hue='Had_Affair', kind='count', palette='coolwarm')
```

```python
sns.factorplot('educ', data=df, hue='Had_Affair', kind='count', palette='coolwarm')
```

```python
occ_dummies = pd.get_dummies(df['occupation'])
```

```python
hus_occ_dummies = pd.get_dummies(df['occupation_husb'])
```

```python
occ_dummies.head()
```

```python
occ_dummies.columns = ['occ1', 'occ2', 'occ3', 'occ4', 'occ5', 'occ6']
```

```python
hus_occ_dummies.columns = ['hocc1', 'hocc2', 'hocc3', 'occ4', 'hocc5', 'hocc6']
```

```python
X = df.drop(['occupation', 'occupation_husb', 'Had_Affair'], axis=1)
```

```python
X.head()
```

```python
dummies = pd.concat([occ_dummies, hus_occ_dummies], axis =1)
```

```python
dummies
```

```python
X = pd.concat([X, dummies], axis=1)
```

```python
X.head()
```

```python
Y = df.Had_Affair
```

```python
Y.head()
```

```python
X.head()
```

```python
X = X.drop('occ1', axis=1)
```

```python
X.head()
```

```python
X = X.drop('hocc1', axis=1)
```

```python
X.head()
```

```python
X = X.drop('affairs', axis=1)
```

```python
Y.head()
```

```python
Y = np.ravel(Y)
```

```python
Y
```

```python
log_model = LogisticRegression()

log_model.fit(X, Y)

# accuracy 
log_model.score(X,Y)
```

```python
Y.mean()
```

```python
coeff_df = DataFrame(list(zip(X.columns, np.transpose(log_model.coef_))))
```

```python
coeff_df
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y)
```

```python
log_model2 = LogisticRegression()

log_model2.fit(X_train, Y_train)
```

```python
class_predict = log_model2.predict(X_test)
```

```python
print (metrics.accuracy_score(Y_test, class_predict))
```
